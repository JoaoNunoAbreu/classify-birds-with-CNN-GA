# Visualiza√ß√£o do dataset

Podemos ver no gr√°fico abaixo a distribui√ß√£o do dataset pelas pastas, tendo o n√∫mero de imagens de cada dataset associado √† esp√©cie.

```python
import os
import matplotlib.pyplot as plt
classes = sorted(os.listdir(train_directory))

counter = [len(os.listdir(train_directory + "/" + i)) for i in classes]

counter_less_than_100 = [("N¬∫ imagens:",counter[i],"Esp√©cie:",classes[i]) for i in range(len(counter)) if(counter[i] < 100)]
for i in counter_less_than_100:
  print(i)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.xlabel("N√∫mero da Esp√©cie (ordem alfab√©tica)")
plt.ylabel('N√∫mero de imagens')
ax.bar([i for i in range(250)],counter)
plt.show()
```
